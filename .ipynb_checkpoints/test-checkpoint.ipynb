```python
import pandas as pd

pf_exa = pd.read_csv('./大學校院招生報考人數.csv')
pf_exa.drop(pf_exa.columns[[2,3,4]], axis=1, inplace=True)
pf_exa
```

```python
pf_adm = pd.read_csv('./大學校院招生錄取人數.csv')
pf_adm.drop(pf_adm.columns[[2,3,4]], axis=1, inplace=True)
pf_adm
```

```python
per_tot_exa = pf_exa.groupby('Period').apply(lambda x:x[ (x['Category1Title']=='總計') & (x['Category2Title']=='個人申請') ])
per_tot_adm = pf_adm.groupby('Period').apply(lambda x:x[ (x['Category1Title']=='總計') & (x['Category2Title']=='個人申請') ])
```

```python
per_tot = per_tot_exa.rename(columns={'Val':'報考人數'})
per_tot['錄取人數'] = per_tot_adm['Val']
per_tot['錄取率'] = (per_tot['錄取人數']/per_tot['報考人數']).apply(lambda x: format(x, '.2%'))
per_tot.drop(per_tot.columns[[0,1,2]], axis=1, inplace=True )
per_tot = per_tot.droplevel(1, axis=0)

per_tot
```
