# HW1 - 高等中學應屆畢業生升學狀況分析

資料來源：https://data.gov.tw/dataset/9625

```python
import pandas as pd

# 讀取原始資料
df_109 = pd.read_csv('./109_1_2_1.csv')
df_109.set_index(['學程別','設立別', '性別'], inplace=True)

# 將比率和人數分成不同 DataFrame
df_109_rate = df_109[df_109.filter(regex='比率').columns]
df_109_num = df_109[df_109.filter(regex='人數').columns]

df_109.head(10)
```

### Q1: 這份資料中各學程別、設立別、性別的畢業生共有多少？

```python
#學程別
pd.DataFrame( df_109.groupby('學程別')['畢業生人數'].sum() )
```

```python
#設立別
pd.DataFrame( df_109.groupby('設立別')['畢業生人數'].sum() )
```

```python
#性別
pd.DataFrame( df_109.groupby('性別')['畢業生人數'].sum() )
```

### Q2: 各行(row)的比率相加是否確實等於升學率？

```python
res = pd.DataFrame({'sum' : df_109_rate.sum(axis=1)})
res['升學率'] = df_109['升學率']
res['差值'] = res['sum']-res['升學率']

res
```

### Q3: 各學程別、設立別、性別的升學率分別是多少？

```python
st_num = pd.DataFrame({'畢業生人數':df_109['畢業生人數'], '升學人數':round(df_109['畢業生人數'] * df_109['升學率'])})

# 學程別
res = st_num.groupby('學程別').sum()
res['升學率'] = round(res['升學人數'] / res['畢業生人數'], 2)
res.sort_values(by='升學率', ascending=False)
```

```python
# 設立別
res = st_num.groupby('設立別').sum()
res['升學率'] = round(res['升學人數'] / res['畢業生人數'], 2)
res.sort_values(by='升學率', ascending=False)
```

```python
# 性別
res = st_num.groupby('性別').sum()
res['升學率'] = round(res['升學人數'] / res['畢業生人數'], 2)
res.sort_values(by='升學率', ascending=False)
```

### Q4: 各個大專學制中應屆畢業生的男女比例

```python
group_res = df_109_num.groupby('性別').sum()
res = pd.DataFrame({'畢業生人數':group_res.sum(), '女性比例' : round(group_res.iloc[0]/group_res.sum(), 2)})
res
```

### Q5: 哪種畢業生最願意讀此大專學制？（哪種畢業生的比率最高）

```python
max_idx = df_109_rate.idxmax(axis=0)
max_rate = df_109_rate.max(axis=0)
res = pd.DataFrame({'最高比率':max_rate, '畢業生類型':max_idx})

# 以畢業生類型當作index去取對應類型畢業生的升學率
res['此類型畢業生升學率'] = res.apply(lambda row: df_109.xs(row['畢業生類型'])['升學率'], axis=1)
res['占比'] = round(res['最高比率']/res['此類型畢業生升學率'], 2)

res
```

### Q6: 高中設立別與大專院校設立別的關係

```python
group_res = df_109_num.groupby('設立別').sum()
res = pd.DataFrame({'公立大專人數':group_res.filter(regex='公立').sum(axis=1),'私立大專人數':group_res.filter(regex='私立').sum(axis=1)})

res.insert(0, '總人數', group_res['畢業生人數'])
res.index += '高中'
res['公立大專比率'] = round(res['公立大專人數'] / res['畢業生人數'], 2)
res['私立大專比率'] = round(res['私立大專人數'] / group_res['畢業生人數'], 2)
res = res.drop(columns=['公立大專人數', '私立大專人數'])
res
```

### Q7: 各大專學制學生來源分佈

#### 依學程別

```python
res = df_109_num.drop(columns='畢業生人數').groupby('學程別').sum().transpose()

res.index = pd.Series(res.index).apply(lambda s: s[:-2]) #刪除'人數'二字
res = res.add_suffix('比率')


res.columns.name = None
total_st = res.sum(axis=1)
res = res.div(total_st, axis=0).apply(lambda x: round(x, 2))
res.insert(0, '總人數', total_st)
res
```

#### 伊設立別

```python
res = df_109_num.drop(columns='畢業生人數').groupby('設立別').sum().transpose()

res.index = pd.Series(res.index).apply(lambda s: s[:-2]) #刪除'人數'二字
res = res.add_suffix('高中比率')


res.columns.name = None
total_st = res.sum(axis=1)
res = res.div(total_st, axis=0).apply(lambda x: round(x, 2))
res.insert(0, '總人數', total_st)
res
```

#### 依性別（同Q4）

```python
res = df_109_num.drop(columns='畢業生人數').groupby('性別').sum().transpose()

res.index = pd.Series(res.index).apply(lambda s: s[:-2]) #刪除'人數'二字
res = res.add_suffix('性比率')


res.columns.name = None
total_st = res.sum(axis=1)
res = res.div(total_st, axis=0).apply(lambda x: round(x, 2))
res.insert(0, '總人數', total_st)
res
```
